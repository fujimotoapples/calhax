# WAVES

```python
with open('wave.json', "r") as f:
    wave_dictionary=json.load(f)
wave_dictionary
```

```python
#this converts unix time to standard time for ease of reading
def timemaker(dictionary):
    return(time.strftime("%a, %d %b %Y %H:%M:%S +0000",time.localtime(dictionary["timestamp"])))

#wavemaker takes in the dictionary and converts into dataframe with relevant info for each timestamp
def wavemaker(dictionary):
    wave=pd.DataFrame.from_dict(dictionary['swells'],orient='columns').drop(columns=["directionMin"])
    return wave

#the dictionary is at the timestamp level. Same dictionary works on both functions

wave_labels=np.array(['height1', 'height2', 'height3', 'height4', 'height5', 'height6',
        'period1', 'period2', 'period3', 'period4', 'period5', 'period6',
        'direction1', 'direction2', 'direction3', 'direction4', 'direction5',
        'direction6','optimal_wave1','optimal_wave2','optimal_wave3','optimal_wave4',
                      'optimal_wave5','optimal_wave6'])
#wave_table = pd.DataFrame(columns = wave_labels)
#row = pd.DataFrame(columns = wave_labels)
```

```python
time_0 = wave_dictionary['data']['wave'][0]
wave_table = wavemaker(time_0).melt().transpose().drop('variable')
wave_table.columns = wave_labels
wave_table['timestamp'] = time_0['timestamp']
wave_table
#wave_table = wave_table[['timestamp','height1', 'height2', 'height3', 'height4', 'height5', 'height6',
#        'period1', 'period2', 'period3', 'period4', 'period5', 'period6',
#        'direction1', 'direction2', 'direction3', 'direction4', 'direction5',
#        'direction6','optimal_wave']]
#wave_table
```

```python

for i in range(1,len(wave_dictionary['data']['wave'])):    
    time = wave_dictionary['data']['wave'][1]
    row = wavemaker(time).melt().transpose().drop('variable')
    row.columns = wave_labels
    row['timestamp'] = time['timestamp']
    row = row[['timestamp','height1', 'height2', 'height3', 'height4', 'height5', 'height6',
        'period1', 'period2', 'period3', 'period4', 'period5', 'period6',
        'direction1', 'direction2', 'direction3', 'direction4', 'direction5',
        'direction6','optimal_wave1','optimal_wave2','optimal_wave3','optimal_wave4',
                      'optimal_wave5','optimal_wave6']]
    wave_table = pd.concat([wave_table,row], join = 'inner')
```

```python
wave_table = wave_table.reset_index()
wave_table = wave_table.drop(columns = ['index'])
```

# WIND

```python
with open('wind.json',"r") as w:
    wind_dictionary=json.load(w)
wind_dictionary
```

```python
def windmaker(dictionary):
    wind=pd.DataFrame.from_dict(dictionary,orient='columns').drop(columns=['utcOffset',])
    return wind
wind_dict=wind_dictionary['data']['wind']
wind_table=windmaker(wind_dict)
```

```python
wind_table
```

# TIDE


```python
with open('tide.json',"r") as w:
    tide_dictionary=json.load(w)

def tidemaker(dictionary):
    tide=pd.DataFrame.from_dict(dictionary,orient='columns').drop(columns=['utcOffset'])
    return tide
at_a_time=tide_dictionary['data']['tides']
tide_table=tidemaker(at_a_time)
def rate_tide(height):
    if height <= 3 and height >= 2.5:
        return 3
    elif height < 2:
        return 1
    else:
        return 0
tide_table['tideOptimal'] = tide_table.height.apply(rate_tide)
tide_table
```

# MASTERbater


```python
master = pd.merge(wave_table,wind_table, how='inner', on= 'timestamp')
```

```python
master = pd.merge(master,tide_table, how='inner', on= 'timestamp')
```

```python
master['score'] = master['tideOptimal'] + master['optimalScore'] + master['optimal_wave6'] + master['optimal_wave5'] + master['optimal_wave4'] + master['optimal_wave3'] +master['optimal_wave2'] + master['optimal_wave1']
```
